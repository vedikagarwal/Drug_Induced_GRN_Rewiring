```python
import polars as pl

file_path = 'lioness_annot.txt'

df_disease = pl.read_csv(
    file_path,
    separator='\t',  
    has_header=True
)

df_disease.head()
```

```python
df_subset_cancer_1 = pl.read_csv(
    'subset_cancer_1.csv',
    has_header=True
)
df_subset_cancer_1.head()
df_subset_cancer_2 = pl.read_csv(
    'subset_cancer_2.csv',
    has_header=True
)
df_subset_cancer_2.head()
df_subset_cancer = pl.concat([df_subset_cancer_1, df_subset_cancer_2])

name_list = df_subset_cancer[:, 0].to_list()
```

```python
name_list_split = [name.split() for name in name_list]
sample_names = [name[0] for name in name_list_split[:]]  
```

### Make the number of rows the same

```python
intersecting_samples = list(set(sample_names) & set(df_disease.columns))
df_disease_samples = df_disease.select(['TF', 'gene'] + intersecting_samples)
df_disease_samples
```

```python
df_control = pl.read_csv(
    'Colon_Sigmoid_AllSamples.csv',
    has_header=True
)           

df_control = df_control.with_columns([
    pl.col(df_control.columns[0])
      .str.split_exact("::", 1)
      .alias("TF_gene")
]).unnest("TF_gene").rename({"field_0": "TF", "field_1": "gene"})

cols = ['TF', 'gene'] + [col for col in df_control.columns if col not in ('TF', 'gene')]
df_control = df_control.select(cols)
df_control.head()

df_control = df_control.drop(df_control.columns[2])
df_control.head()
```

```python
tf_intersection = set(df_disease['TF'].to_list()) & set(df_control['TF'].to_list())
print(f"Number of intersecting TFs: {len(tf_intersection)}")
print(list(tf_intersection)[:10])  # print first 10 as example
```

```python
df_disease = df_disease_samples
```

```python
df_control.head()
```

```python
df_control.shape
```

```python
sample_col = df_control.columns[2]  

adj_matrix = df_control.select(['TF', 'gene', sample_col]).to_pandas().pivot(
    index='TF', columns='gene', values=sample_col
)

adj_matrix
```

```python
sample_col_disease = df_disease.columns[2]  

adj_matrix_disease = df_disease.select(['TF', 'gene', sample_col_disease]).to_pandas().pivot(
    index='TF', columns='gene', values=sample_col_disease
)

adj_matrix_disease
```

```python
import mygene
ensembl_ids = adj_matrix.columns

mg = mygene.MyGeneInfo()
query_results = mg.querymany(
    ensembl_ids,
    scopes='ensembl.gene',
    fields='symbol',
    species='human'
)
id_to_symbol = {res['query']: res.get('symbol', res['query']) for res in query_results}

adj_matrix.rename(columns=id_to_symbol, inplace=True)
```

```python
intersecting_genes = set(adj_matrix.columns) & set(adj_matrix_disease.columns)
print(f"Number of intersecting genes: {len(intersecting_genes)}")
print(list(intersecting_genes)[:10])  
```

```python
common_names = set(adj_matrix.columns) & set(adj_matrix.index)
print(f"Number of common names: {len(common_names)}")
print(list(common_names))
```

```python
common_tfs = set(adj_matrix.index) & set(adj_matrix_disease.index)
print(f"Number of TFs in common: {len(common_tfs)}")
print(list(common_tfs)[:10])  
```

```python
df_disease_filtered = df_disease.filter(pl.col("TF").is_in(common_tfs))
df_control_filtered = df_control.filter(pl.col("TF").is_in(common_tfs))
```

```python
df_disease_filtered.shape
df_control_filtered.shape
```

```python
df_control_filtered = df_control_filtered.with_columns(
    pl.col("gene").replace(id_to_symbol).alias("gene")
)
```

```python
df_control_filtered
```

```python
genes_control = set(df_control_filtered['gene'].to_list())
genes_disease = set(df_disease_filtered['gene'].to_list())
genes_in_common = genes_control & genes_disease
print(f"Number of genes in common: {len(genes_in_common)}")
print(list(genes_in_common)[:10])  
```

```python
df_control_filtered_genes = df_control_filtered.filter(pl.col("gene").is_in(genes_in_common))
df_disease_filtered_genes = df_disease_filtered.filter(pl.col("gene").is_in(genes_in_common))
```

```python
df_control_filtered_genes.shape, df_disease_filtered_genes.shape
```

```python
df_control_joint = df_control_filtered_genes.join(
    df_disease_filtered_genes, 
    on=['TF', 'gene'], 
    how='semi'
)
df_disease_joint = df_disease_filtered_genes.join(
    df_control_filtered_genes, 
    on=['TF', 'gene'], 
    how='semi'
)

print(df_control_joint.shape, df_disease_joint.shape)
```

```python
df_c_unique = df_control_joint.group_by(['TF', 'gene']).mean()
df_d_unique = df_disease_joint.group_by(['TF', 'gene']).mean()

df_c_final = df_c_unique.join(df_d_unique, on=['TF', 'gene'], how='semi').sort(['TF', 'gene'])
df_d_final = df_d_unique.join(df_c_unique, on=['TF', 'gene'], how='semi').sort(['TF', 'gene'])

print(df_c_final.shape, df_d_final.shape)
```

```python
df_c_final.write_csv("df_c_final.csv")
df_d_final.write_csv("df_d_final.csv")

df_control_final = pl.read_csv("df_c_final.csv")
df_disease_final = pl.read_csv("df_d_final.csv")
```
